Assignment; Implement the remaining 3 methods of handling missing values

__Interpolation__  

 -Linear Interpolation  
 -Polynomial Interpolation  
 -Spline Interpolation  
 -Forward Fill Interpolation  
 -Backward Fill Interpolation  

Based on the nature of the dataset, spline interpolation might not be of much help.

```python
#Using Linear Interpolation
student_data = student_data.interpolate(method = 'linear')
student_data
```

```python
#Using Polynomial Interpolation
student_data = student_data.interpolate(method='polynomial', order = 2)
student_data
```

```python
#Using Spline Interpolation
student_data = student_data.interpolate(method='spline', order = 3)
student_data
```

__Filling__  
- we have the forward fill and backward fill which basically inputs the next or previous value respectively.

```python
#Using forward fill
student_data = student_data.fillna(method='ffill')
student_data
```

```python
#Using backward fill
student_data = student_data.fillna(method='bfill')
student_data
```

__Imputation__  
- This is where we use methods such as mean, median and mode to improvise for the missing values
- There is also the KNN imputation which is mainly used for numeric columns and in complex data.
- And there is iterative learning which is mainly Machine Learning-based

```python
#Using mean
student_data.fillna(student_data.mean(), inplace = True)
student_data
```

```python
#Using median
student_data.fillna(student_data.median(), inplace = True)
student_data
```

```python
#Using mode
student_data.fillna(student_data.mode(), inplace = True)
student_data
```
